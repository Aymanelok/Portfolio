# 1. Import Libraries

````python
import numpy as np
import pandas as pd
import math
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
````

# 2. Load an Online Corpus

In this lab we will use a **subset of the 20 Newsgroups dataset**, a classical NLP dataset.

````python
from sklearn.datasets import fetch_20newsgroups

dataset = fetch_20newsgroups(
    remove=('headers','footers','quotes'),
    categories=['sci.med','sci.space','comp.graphics']
)

documents = dataset.data[:200]

print(f"Loaded {len(documents)} documents.")
````

# 3. Simple Tokenization

````python
tokenized_docs = [doc.lower().split() for doc in documents]

print("Tokens from first document:")
print(tokenized_docs[0][:30])
````

# 4. CHALLENGE - Implement TF

Implement a function:

```
compute_tf(document)
```

that returns a dictionary:

```
{word: tf_score}
```

Where:

$$TF(t,d) = \frac{\text{count}(t,d)}{\text{total words in document}}$$

````python
def compute_tf(document):
    tf_scores = {}
    word_counts = Counter(document)
    total_words = len(document)

    if total_words == 0:
        return tf_scores # Avoid division by zero for empty documents

    for word, count in word_counts.items():
        tf_scores[word] = count / total_words
    return tf_scores
````

# 5. CHALLENGE - Implement IDF

Implement:

```
compute_idf(corpus)
```

Where:

$$IDF(t) = \log\left(\frac{N}{df(t)}\right)$$

- N = number of documents
- df(t) = number of documents containing term t

````python
def compute_idf(corpus):
    import math
    N = len(corpus)
    idf_scores = {}

    # Get all unique words in the entire corpus
    all_words = set(word for doc in corpus for word in doc)

    for word in all_words:
        # Count how many documents contain this word
        df_t = sum(1 for doc in corpus if word in set(doc))

        # Standard IDF formula
        idf_scores[word] = math.log(N / df_t)

    return idf_scores
````

# 6. CHALLENGE - Implement TF‑IDF

Create:

```
compute_tfidf(document, idf_scores)
```

that returns:

```
{word: tfidf_score}
```

````python
def compute_tfidf(document, idf_scores):
    tfidf_scores = {}
    tf_scores = compute_tf(document)

    for word, tf_val in tf_scores.items():
        # Use idf_scores.get(word, 0) to handle words not in the IDF dictionary
        tfidf_scores[word] = tf_val * idf_scores.get(word, 0)

    return tfidf_scores
````

# 7. Test Your Implementation

Compute TF‑IDF for the first document.

````python
# Compute IDF for the entire corpus
idf_scores = compute_idf(tokenized_docs)

# Compute TF-IDF for the first document
tfidf_doc = compute_tfidf(tokenized_docs[0], idf_scores)

# Sort and show top 20 words
top_20 = sorted(tfidf_doc.items(), key=lambda x: x[1], reverse=True)[:20]
for word, score in top_20:
    print(f"{word:20} : {score:.5f}")
````
